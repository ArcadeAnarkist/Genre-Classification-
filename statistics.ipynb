```python
import numpy as np
```

# Samplesize

```python
p = 0.54
n = 1.96**2 * p * (1 - p) / (0.025)**2
print(f"RBF : {int(n)}")

p = 0.68
n = 1.96**2 * p * (1 - p) / (0.025)**2
print(f"RF: {int(n)}")

p = 0.59
n = 1.96**2 * p * (1 - p) / (0.025)**2
print(f"NN: {int(n)}")

print('Biggest samplesize needed is 1526')
```

# Confidence interval  

```python
n = 1526


p = 0.54
c_low = p - 1.96 * np.sqrt((p * (1 - p)) / n)
c_high = p + 1.96 * np.sqrt((p * (1 - p)) / n)
print(f"RBF = 95% CI: [{c_low:.3f}, {c_high:.3f}]")
print(c_high - c_low)

p = 0.68
c_low = p - 1.96 * np.sqrt((p * (1 - p)) / n)
c_high = p + 1.96 * np.sqrt((p * (1 - p)) / n)
print(f"RF = 95% CI: [{c_low:.3f}, {c_high:.3f}]")
print(c_high - c_low)

p = 0.59
c_low = p - 1.96 * np.sqrt((p * (1 - p)) / n)
c_high = p + 1.96 * np.sqrt((p * (1 - p)) / n)
print(f"NN = 95% CI: [{c_low:.3f}, {c_high:.3f}]")
print(c_high - c_low)
```
